```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

print("Libaries Loaded")
```

```python
def loadData(path, val_name):
    #Load data 
    df = pd.read_csv(path)
    #Melt data
    df = df.melt(id_vars=['Country Name', 'Country Code'], value_vars=df.columns[4:65], var_name='Year',value_name=val_name)
    return(df)

#Load CO2_data
co2Path="D:/Documents/UT-Austin/Markets for Electricity/Data for Project/CO2PerCapita_Data_Extract_From_World_Development_Indicators/23368d8d-9a3f-49fc-b4aa-34e0c2eca326_Data.csv"
co2_data = loadData(co2Path, 'CO2')

#Load gdp per capita data
gdp_path="D:/Documents/UT-Austin/Markets for Electricity/Data for Project/GDPPerCapita_Data_Extract_From_World_Development_Indicators/3dc335a3-ec38-4137-a9ac-a9668d3633da_Data.csv"
gdp_data = loadData(gdp_path, 'GDP')

#Load Gini Data
gini_df=loadData("D:/Documents/UT-Austin/Markets for Electricity/Data for Project/GINI_Data_Extract_From_World_Development_Indicators/1883979a-1c2d-4c5d-a715-b8ce106f56bf_Data.csv", 'GINI')
gini_df["Year"] = pd.to_numeric(gini_df["Year"].map(lambda x: x[0:4]))

#Load Alt Electricity
AltElectricPath="D:/Documents/UT-Austin\Markets for Electricity/Data for Project/PercentAltEnergy_Data_Extract_From_World_Development_Indicators/6c988d8e-99a0-4039-8fcc-20609f39db22_Data.csv"
alt_df = loadData(AltElectricPath, 'Alternative Energy')
alt_df["Year"] = pd.to_numeric(alt_df["Year"].map(lambda x: x[0:4]))



#Merge data
df = co2_data.merge(gdp_data, left_on=['Country Name', 'Country Code', 'Year'], right_on=['Country Name', 'Country Code', 'Year'])
df = df.replace('..',np.NaN)
df = df.dropna()
df["GDP"] = pd.to_numeric(df["GDP"])
df["CO2"] = pd.to_numeric(df["CO2"])
df["Year"] = pd.to_numeric(df['Year'].map(lambda x: x[0:4]))
df["GDPSquare"] = df["GDP"]**2
df["GDPCube"] = df["GDP"]**3
#Merge gini data afterwards so that we don't drop data for years where there's no gini index
#df = df.merge(gini_df, left_on=['Country Name', 'Country Code', 'Year'], right_on=['Country Name', 'Country Code', 'Year'])
#Merge alternative energy df and make the value numeric
df = df.merge(alt_df, left_on=['Country Name', 'Country Code', 'Year'], right_on=['Country Name', 'Country Code', 'Year'])
df=df.replace('..',np.NaN)
df['Alternative Energy'] = pd.to_numeric(df['Alternative Energy'])
df['AltTimesGDP'] = df['Alternative Energy']*df['GDP']
df['AltTimesGDPSquare'] = df['Alternative Energy']*df['GDPSquare']
df['AltTimesGDPCube'] = df['Alternative Energy']*df['GDPCube']

df.head()
```

```python
country_codes = [
    'MAR', #Morocco
    'MEX', #Mexico
    'CHN', #China
    'KOR', #Korea
    'IND', #India
    'SEN', #Senegal
    'FRA', #France
    'SGP', #Singapore
    'RWA', #Rwanda
    'PAK', #Pakistan
    'CIV', #Ivory Coast
    'BRA', #Brazil
    'JPN', #Japan
    'PER', #Peru
    'IRL', #Ireland
    'CAN', #Canada
    'TUR', #Turkey
    'AUS', #Australia
    'NGA', #Nigeria
    'USA', #United States
    'GRC', #Greece
    'FIN', #Finland
    'KEN', #Kenya
]
regions = [
    'AFE', #Africa East & South
    'AFW', #Africa West and Central
    'ARB', #Arab World
    'CEB', #Central Europe and the Baltics
    'EAS', #East Asia and Pacific
    'ECS', #Europe & Central Asia
    'LCN', #Latin America and Carribbean
    'MEA', #Middle East & North Africa
    'NAC', #North America
    'SAS', #South Asia
    'SSF', #Sub-Saharan Africa
    'WLD', #World
]
demographics = [
    'FCS', #Fragile and conflict affected
    'HPC', #Heavily indebted poor countries
    'HIC', #High income
    'IBD', #IBRD Only
    'IBT', #IDA & IBRD Total
    'IDX', #IDA only
    'LDC', #Least developed countries
    'LMY', #Low / middle income
    'LIC', #Low income
    'LMC', #Lower middle income
    'MIC', #Middle Income
    'OED', #OECD Members
    'SST', #Small states
    'UMC', #Upper middle income
]
```

```python
#Create dataframes based on countries, regions, and demographics
country_df = df[df['Country Code'].isin(country_codes)]
region_df = df[df['Country Code'].isin(regions)]
demo_df = df[df['Country Code'].isin(demographics)]

#Check they all have same amount of data available
country_df['Country Name'].value_counts()
```

```python
#Function for plotting the GDP vs CO2 data and adding a trendline on it
def PlotAndModel(df, title):

    #Plotting CO2 per capita against GDP per capita
    plt.scatter(df['GDP'],df['CO2'])

    #Quadratic model of the above
    model = np.poly1d(np.polyfit(df['GDP'],df['CO2'], 2))
    polyline = np.linspace(min(df['GDP']),max(df['GDP']), 200)
    plt.plot(polyline, model(polyline), color="#6D34D6", linestyle='dashed')
    
    plt.title(title)
    plt.xlabel("GDP Per Capita ($USD)")
    plt.ylabel("CO2 Emissions Per Capita (Metric Tons)")
    plt.show()
    
PlotAndModel(country_df, 'All Countries')
PlotAndModel(region_df, 'All Regions')
PlotAndModel(demo_df, 'By World Bank Demographic')
```

```python
#Plots by country
def PlotByCountry(df):
    for country in df['Country Name'].unique():
        temp_df = df[df['Country Name'] == country]

        plt.scatter(temp_df['GDP'], temp_df['CO2'])

        #Quadratic model of the above
        model = np.poly1d(np.polyfit(temp_df['GDP'], temp_df['CO2'], 2))
        polyline = np.linspace(min(temp_df['GDP']), max(temp_df['GDP']), 200)
        plt.plot(polyline, model(polyline), color="#6D34D6", linestyle='dashed')

        plt.xlabel("GDP Per Capita ($USD)")
        plt.ylabel("CO2 Emissions Per Capita (Metric Tons)")
        plt.title(country)
        plt.show()

PlotByCountry(country_df)
PlotByCountry(region_df)
PlotByCountry(demo_df)
```

```python
#Plots by year
def YearPlot(df):
    for year in df['Year'].unique():
        temp_df = df[df['Year'] == year]

        plt.scatter(temp_df['GDP'], temp_df['CO2'])

        #Quadratic model of the above
        model = np.poly1d(np.polyfit(temp_df['GDP'], temp_df['CO2'], 2))
        polyline = np.linspace(min(temp_df['GDP']), max(temp_df['GDP']), 200)
        plt.plot(polyline, model(polyline), color="#6D34D6", linestyle='dashed')

        plt.title(year)
        plt.show()
        
YearPlot(country_df)
```

```python
#Run regression
def Regress(df):
    regress = sm.OLS(df['CO2'], df[['GDP','GDPSquare', 'Year']]).fit()
    return(regress.summary())

Regress(country_df)
#Regress(region_df)
#Regress(demo_df)
```

```python
#Run instrument variable analysis with year as an instrument on alternative energy
from statsmodels.sandbox.regression.gmm import IV2SLS

two_iv_results = IV2SLS(country_df['CO2'], country_df[['GDP','GDPSquare', 'Alternative Energy']], instrument = country_df['Year']).fit()
two_iv_results.summary()
```

```python
#Attempt at Fixed Effects Regression
# Import model
from linearmodels.panel import PanelOLS

#Set Index
country_df = country_df.set_index(['Country Code', 'Year'])
```

```python
# Model
m = PanelOLS(dependent=country_df['CO2'],
             exog=country_df[['GDP'
                              ,'GDPSquare'
                              ,'Alternative Energy'
                             ]],
             entity_effects=True,
             time_effects=False,
             )
m.fit(cov_type='clustered', cluster_entity=True)
```
